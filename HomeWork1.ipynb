```python
vacs = list(csv.DictReader(open('vacancy.csv', encoding='utf-8')))
```

```python
# Сколько вакансий, которые вам нравятся?
tmp = 0
vac_lst = []
for i in range(len(vacs)):
    if vacs[i]['vactitle'].lower().find('аналитик')>=0:
        print(vacs[i]['id'], '   ', vacs[i]['vactitle'])
        vac_lst.append(vacs[i])
        tmp+=1
print('Итого: ', tmp)
```

```python
# Насколько свежие эти вакансии?
dates_lst = []
for i in range (len(vac_lst)):
    dates_lst.append(vac_lst[i]['updated_at'])
max(dates_lst)
```

```python
# Сколько вакансий с позициями на которых вы работаете?
tmp = 0
vac_lst = []
for i in range(len(vacs)):
    if vacs[i]['vactitle'].lower().find('главный инженер по разработке')>=0:
        print(vacs[i]['id'], '   ', vacs[i]['vactitle'])
        vac_lst.append(vacs[i])
        tmp+=1
print('Итого: ', tmp)
```

```python
# Сколько вакансий для аналатика данных?
tmp = 0
vac_lst = []
for i in range(len(vacs)):
    name = vacs[i]['vactitle'].lower()
    if (name.find('аналитик данных')>=0)|(name.find('data analyst')>=0):
        print(vacs[i]['id'], '   ', vacs[i]['vactitle'])
        vac_lst.append(vacs[i])
        tmp+=1
print('Итого: ', tmp)
```

```python
# Сколько вакансий для аналитика данных с использованием Python?
tmp = 0
vac_lst = []
for i in range(len(vacs)):
    name = vacs[i]['vactitle'].lower()
    if (((name.find('аналитик данных')>=0)|(name.find('data analyst')>=0))&
            (vacs[i]['vacdescription'].lower().find('python')>=0)):
        print(vacs[i]['id'], '   ', vacs[i]['vactitle'])
        vac_lst.append(vacs[i])
        tmp+=1
print('Итого: ', tmp)
```
